**Задание 4**

Реализуйте функцию на языке Python, принимающую на вход n и возвращающую значение следующей суммы: 

```python
def ln_sum(n):
    if n < 1:
        return 0
    return log(n) + ln_sum(n-1)

ln_sum(5)
```

**Задание 5**

Реализуйте функцию на языке Python, принимающую на вход n и возвращающую значение следующего выражения: 

```python
def ln_product(n):
    product = 1
    for i in range(1, n+1):
        product = product*i
    return log(product)

ln_product(5)
```

Результатами первых двух заданий должны быть функции, которые принимают на вход число, а на выход возвращают посчитанное выражение - логарифм произведения всех чисел от 1 до указанного числа.
При этом в первом случае мы считаем логарифм произведения чисел как сумму логарифмов, а во втором - мы сперва находим произведение чисел и только потом используем функцию логарифма.

Ниже вы найдете код, который создает ползунки, с помощью которых можно варьировать коэффициенты a, b и c, подбирая нужный вид параболы, а также рисует набор точек. <br><br>

### 11 - Функции потерь

```python
from sympy import *
from sympy.plotting import plot
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
def print_points_ands_function(sympy_function):
    def function(x_): return float(sympy_function.subs(x, x_))
    
    points_X = np.array([-3, -2, -1, 1, 2, 3])
    points_Y = np.array([15.0, 8.0, 13.0, 10.0, 15.0, 30.0])
    plt.xlim(-25, 7.5)
    plt.ylim(-1, 40)
    
    plt.scatter(points_X, points_Y, c='r')
    x_range = np.linspace(plt.xlim()[0], plt.xlim()[1], num=100)
    function_Y = [function(x_) for x_ in x_range]
    plt.plot(x_range, function_Y, 'b')
    plt.show()
    
    MSE = sum([(points_Y[i] - function(points_X[i]))**2 for i in range(len(points_Y))]) / len(points_Y)
    print(f'MSE = {MSE}')
```

### 12 - Проверка на четность/нечетность

Раньше чтобы проверить функцию на четность/нечетность, мы подставляли в subs() противоположные числа и проверяли, что значение 
функции не изменилось или стало противоположным.

Теперь мы можем подставить в функцию -х вместо х, и если от этого функция не изменится, значит это будет выполняться для любых 
двух противоположных чисел.

```python
# Например, х2 - четная функция:

f = x**2
f
```

```python
# Подставив вместо х -х получили ту же самую функцию:

f.subs(x, -x)

# Подстановка противоположных значений дает тот же результат, значит функция четная
```

```python
# А вот такая функция является нечетной:

f = x**3 + 3*x
f
```

```python
# Потому что если мы подставим в нее -х, получим противоположную функцию:

f.subs(x, -x)
```

А для функций, которые не являются ни четными, ни нечетными, ни того же самого, ни противоположного значения не получится.
