### Задача 0 - Учетная запись

Представим себе такую ситуацию - мы администрируем сеть в какой-нибудь компании. И для каждого нового сотрудника нам необходимо создать учетную запись с именем и паролем, которые по умолчанию будут Admin и qwerty. А так же там будет флаг, который будет означать забанен пользователь в системе или нет. По умолчанию он будет False. 

У нас есть 3 реализации такой задачи.

![title](Классы1.jpg)

```python
# Чтобы создать и работать со своим объектом, его нужно сначала описать
# и это формальное описание объекта называется классом
# Стиль CamelCase используется преимущественно для имен классов
# В атрибутах класса хранится состояние этого класса

class User:
    user_name = 'Admin'    # статические атрибуты класса
    password = 'qwerty'
    is_banned = False
    
# Когда мы описали нашего пользователя, его необходимо создать
# А точнее, взять переменную и присвоить ей наш класс:

user_1 = User()   # экземпляр (объект) класса User (instance)
print(user_1)
print(type(100))
print(type(user_1))
```

```python
# Конечно же, каждый пользователь потом захочет поменять свой логин и пароль
# то есть статические атрибуты нашего класса
# И для получения доступа к этим атрибутам, необходимо воспользоваться знаком ".":

user_1.user_name = 'Tom'
print(user_1.user_name)
```

```python
# Имя мы меняем именно для user_1, сам класс при этом остается абсолютно нетронутым
# Для наглядности создадим еще один экземпляр класса:

user_2 = User()
print(user_2.user_name)
```

```python
# А что если мы хотим добавить в друзья не какое-то случайное имя, а другого пользователя?
# Создадим еще одного юзера:

user_1 = User()
user_2 = User()
user_2.user_name = 'Alina'
user_1.add_friend('Bob')
user_1.add_friend(user_2)
print(user_1.friends)
```

```python
# Мы просто добавили наш экземпляр класса в список
# Чтобы все было по-нормальному, мы можем внутри метода прописать проверку с помощью функции isinstance():

class User:
    user_name = 'Admin'    
    password = 'qwerty'
    is_banned = False
    friends = []         
    
    def print_info(self):
        
        print(
            'Name: {}\nPassword: {}\nBan status: {}'.format(
                self.user_name, self.password, self.is_banned)
        )
        
    def add_friend(self, friend):       
        if isinstance(friend, User):                 # добавили
            self.friends.append(friend.user_name)
        else:
            self.friends.append(friend)
            
user_1 = User()
user_2 = User()
user_2.user_name = 'Alina'
user_1.add_friend('Bob')
user_1.add_friend(user_2)
print(user_1.friends)

# Так мы можем работать и с другимим, созданными нами классами
# Так же не забываем, что в наших методах так же могут использоваться именованные аргументы
# и так же могут возвращаться значения оператором return
```

```python
# Пример из урока:

class Family:
    surname = 'Common family'
    money = 100000
    have_a_house = False
    
    def info(self):
        print(
            'Family name: {}\nFamily funds: {}\nHaving a house: {}\n'.format(
                self.surname, self.money, self.have_a_house
            )
        )
        
    def earn_money(self, amount):
        self.money += amount
        print('Earned {} money! Current value: {}'.format(amount, self.money))
        
    def buy_house(self, house_price, discount=0):
        house_price -= house_price * discount / 100
        if self.money >= house_price:
            self.money -= house_price
            self.have_a_house = True
            print('House purchased! Current money: {}\n'.format(self.money))
        else:
            print('Not enough money!\n')
            
my_family = Family()
my_family.info()

print('Try to buy a house')
my_family.buy_house(10 ** 6)

if not my_family.have_a_house:
    my_family.earn_money(800000)
    print('Try to buy a house again')
    my_family.buy_house(10 ** 6, 10)
    
my_family.info()
```

## 25.4 Конструктор __init__ и работа с несколькими классами

### Задача 0 - Работник

Как мы уже говорили, экземпляров класса может быть достаточно много. И с каждым из них часто приходится работать индивидуально. И конечно же, так же часто их нужно инициализировать с какими-то индивидуальными входными данными. 

В качестве примера возьмем простой класс работника. У этого работника есть 2 статических атрибута - имя и зарплата. И здесь просто метод об отображении информации об этом работнике. Создадим экземпляр класса и проверим:

```python
class Employee:
    name = 'Tom'
    salary = 10000
    
    def print_info(self):
        print('Name: {}\nSalary: {}\n'.format(self.name, self.salary))
        
emp_1 = Employee()
emp_1.print_info()
```

```python
# Но в нашем классе кое-что не так
# Допустим, мы хотим создать еще одного работника с другим именем и зарплатой
# Тогда нам придется сначала создать экземпляр:

emp_2 = Employee()

# Затем переприсвоить имя:

emp_2.name = 'Bob'

# И затем переприсвоить зарплату:

emp_2.salary = 20000

# И только потом мы можем уже полноценно работать с нашим экземпляром:

emp_2.print_info()
```

```python
# Здесь уже понятно, что так дело не пойдет
# Было бы странным, если бы в питоне нельзя было нормально инициализировать объекты
# Именно поэтому при работе с классами почти всегда используется специальный метод,
# который уже прописан в питоне и который отвечает за инициализацию:

def __init__(self):
    
# Этот метод называется конструктором класса
# Двойным подчеркиванием обозначаются магические методы,
# т к они автоматически вызываются в момент инициализации объекта
```

```python
# При инициализации нового работника мы хотим задавать ему имя и зарплату,
# а значит это нужно передать сюда в качестве аргументов:

# А дальше все очень просто - через ключевое слово self мы как бы инициализируем нужные нам атрибуты:

class Employee:
    name = 'Tom'       # здесь обычно объявляют константы для класса или какой-нибудь счетчик
    salary = 10000
    
    def __init__(self, name, salary):         # здесь мы определили атрибуты самому объекту
        self.name = name
        self.salary = salary
    
    def print_info(self):
        print('Name: {}\nSalary: {}\n'.format(self.name, self.salary))
        
emp_1 = Employee()
emp_1.print_info()
```

```python
# Поэтому то, что у нас есть в начале, можем просто убрать:

class Employee:
    
    def __init__(self, name, salary):         
        self.name = name
        self.salary = salary
    
    def print_info(self):
        print('Name: {}\nSalary: {}\n'.format(self.name, self.salary))

# И теперь при создании экземпляра, в скобки достаточно передавать просто название атрибута:

emp_1 = Employee('Tom', 10000)
emp_2 = Employee('Bob', 20000)
emp_1.print_info()
emp_2.print_info()

# Таким образом, с помощью конструктора init мы иниуиализировали 2 атрибута для класса Работник
# Такие атрибуты уже называются динамическими
# Кстати говоря, тут нередко используются и значения по умолчанию, т е именованные аргументы
```

Грядка с картошкой содержит список картошки, которая на ней растёт, и может, собственно, взращивать всю эту картошку, а также предоставлять информацию о зрелости всей картошки на своей территории.

```python
# Пример из урока:

class Potato:   # класс картошки
    
    states = {0: 'Отсутствует', 1: 'Росток', 2: 'Зеленая', 3: 'Зрелая'}     # стадии созревания картошки
    
    def __init__(self, index):
        self.index = index      # по заданию у каждой картошки должен быть номер
        self.state = 0          # стадия зреслости
        
    def grow(self):             # метод для роста картошки
        if self.state < 3:      # сделаем так, чтобы картошка не переросла
            self.state += 1
        self.print_state()
        
    def is_ripe(self):          # метод, который проверяет каждую картошку на зрелость        
        if self.state == 3:
            return True
        return False
            
    def print_state(self):      # метод для вывода зрелости картошки
        
        print('Картошка {} сейчас {}'.format(self.index, Potato.states[self.state]))  # возьмем словарь, к которому обратимся
                                                                                           # через класс
    
    
class PotatoGarden:   # класс грядки
    
    def __init__(self, count):   # передаем в него кол-во посаженной картошки
        
        self.potatoes = [Potato(index) for index in range(1, count + 1)]   # список, в котором будет храниться вся картошка
        
    def grow_all(self):          # метод, который взращивает всю картошку на грядке
        
        print('Картошка прорастает!')
        for i_potato in self.potatoes:  # проходим по списску и для каждой картофелины вызываем метод grow()
            i_potato.grow()
            
    def are_all_ripe(self):       # метод, который проверяет, созрела наша поляна или нет 
        
        if not all([i_potato.is_ripe() for i_potato in self.potatoes]):  # с помощью генерации создаем список из значений True
                                                     # и False, а с помощью функции all() проверяем сразу все значения списка
            print('Картошка еще не созрела!\n')
        else:
            print('Вся картошка созрела. Можно собирать!\n')
            
my_garden = PotatoGarden(5)  # посадим в нашу грядку 5 клубней картошки
my_garden.are_all_ripe()
for _ in range(3):           # с помощью цикла 3 раза взрастим картошку
    my_garden.grow_all()
    my_garden.are_all_ripe() # и так же 3 раза проверим ее готовность
```

```python
import math

class Circle:
    
    def __init__(self, x, y, r):
        self.x = x
        self.y = y
        self.r = r
        self.square = 3,14 * r ** 2
        self.perimeter = 2 * 3,14 * r
        
    def k_size(self, k):
        self.r = self.r * k
        self.square = 3,14 * self.r ** 2
        self.perimeter = 2 * 3,14 * self.r
        
    def intersection(self, circle):
        ab = math.sqrt(((self.x - circle.x) ** 2) + ((self.y - circle.y) ** 2))  # расстояние между центрами
        if self.r + circle.r > ab:
            print('Окружности пересекаются')
        else:
            print('Окружности не пересекаются')
        
circle_1 = Circle(5, 6, 2)
circle_2 = Circle(0, 0, 3)
circle_1.intersection(circle_2)
circle_1.k_size(3)
circle_1.intersection(circle_2)
```

```python
class Potato:   # класс картошки
    
    states = {0: 'Отсутствует', 1: 'Росток', 2: 'Зеленая', 3: 'Зрелая'}     # стадии созревания картошки
    
    def __init__(self, index):
        self.index = index      # по заданию у каждой картошки должен быть номер
        self.state = 0          # стадия зреслости
        
    def grow(self):             # метод для роста картошки
        if self.state < 3:      # сделаем так, чтобы картошка не переросла
            self.state += 1
        self.print_state()
        
    def is_ripe(self):          # метод, который проверяет каждую картошку на зрелость        
        if self.state == 3:
            return True
        return False
            
    def print_state(self):      # метод для вывода зрелости картошки
        
        print('Картошка {} сейчас {}'.format(self.index, Potato.states[self.state]))  # возьмем словарь, к которому обратимся
                                                                                           # через класс
    
    
class PotatoGarden:   # класс грядки
    
    def __init__(self, count):   # передаем в него кол-во посаженной картошки
        
        self.potatoes = [Potato(index) for index in range(1, count + 1)]   # список, в котором будет храниться вся картошка
        
    def grow_all(self):          # метод, который взращивает всю картошку на грядке
        
        print('Картошка прорастает!')
        for i_potato in self.potatoes:  # проходим по списку и для каждой картофелины вызываем метод grow()
            i_potato.grow()
            
    def are_all_ripe(self):       # метод, который проверяет, созрела наша поляна или нет 
        
        if not all([i_potato.is_ripe() for i_potato in self.potatoes]):  # с помощью генерации создаем список из значений True
                                                     # и False, а с помощью функции all() проверяем сразу все значения списка
            print('Картошка еще не созрела!\n')
            return False
        else:
            print('Вся картошка созрела. Можно собирать!\n')
            return True
            
class Gardener:
    
    def __init__(self, name, garden):
        self.name = name
        self.garden = garden
        
    def tend_garden(self):     # ухаживает за грядкой
        if len(self.garden.potatoes) != 0:
            print('{} поливает грядку'.format(self.name))
        
        
    def bring_harvest(self):   # собирает урожай
        if self.garden.are_all_ripe():
            self.garden.potatoes = []
            print('Урожай собран!')
        
            
my_garden = PotatoGarden(5)  # посадим в нашу грядку 5 клубней картошки
my_garden.are_all_ripe()
gardener = Gardener(name='Bob', garden=my_garden)
for _ in range(3):           # с помощью цикла 3 раза взрастим картошку
    my_garden.grow_all()
    gardener.bring_harvest()
    gardener.tend_garden()
```

Логика действий человека определяется следующим образом:

Генерируется число кубика от 1 до 6.

Если сытость < 20, то поесть.

Иначе, если еды в доме < 10, то сходить в магазин.

Иначе, если денег в доме < 50, то работать.

Иначе, если кубик равен 1, то работать.

По такой логике эксперимента человеку надо прожить 365 дней.


Реализуйте такую программу и создайте двух людей, живущих в одном доме. Проверьте работу программы несколько раз. Надеемся, эти люди живы...

```python
from random import randint
from termcolor import cprint


class Man:

    def __init__(self, name):
        self.name = name
        self.fullness = 50
        self.house = None

    def __str__(self):
        return 'Я - {}, сытость {}'.format(
            self.name, self.fullness)

    def eat(self):
        if self.house.food >= 10:
            cprint('{} поел'.format(self.name), color='yellow')
            self.fullness += 10
            self.house.food -= 10
        else:
            cprint('{} нет еды'.format(self.name), color='red')

    def work(self):
        cprint('{} сходил на работу'.format(self.name), color='blue')
        self.house.money += 50
        self.fullness -= 10

    def watch_MTV(self):
        cprint('{} смотрел MTV целый день'.format(self.name), color='green')
        self.fullness -= 10

    def shopping(self):
        if self.house.money >= 50:
            cprint('{} сходил в магазин за едой'.format(self.name), color='magenta')
            self.house.money -= 50
            self.house.food += 50
        else:
            cprint('{} деньги кончились!'.format(self.name), color='red')

    def move_into_the_house(self, house):
        self.house = house
        self.fullness -= 10
        cprint('{} Въехал в дом'.format(self.name), color='cyan')

    def act(self):
        if self.fullness <= 0:
            cprint('{} умер...'.format(self.name), color='red')
            return
        dice = randint(1, 6)
        if self.fullness < 20:
            self.eat()
        elif self.house.food < 10:
            self.shopping()
        elif self.house.money < 50:
            self.work()
        elif dice == 1:
            self.work()
        elif dice == 2:
            self.eat()
        else:
            self.watch_MTV()


class House:

    def __init__(self):
        self.food = 50
        self.money = 0

    def __str__(self):
        return 'В доме еды осталось {}, денег осталось {}'.format(
            self.food, self.money)


citizens = [
    Man(name='Бивис'),
    Man(name='Батхед'),
    Man(name='Кенни'),
]

my_sweet_home = House()
for citizen in citizens:
    citizen.move_into_the_house(house=my_sweet_home)

for day in range(1, 366):
    print('================ день {} =================='.format(day))
    for citizen in citizens:
        citizen.act()
    print('--- в конце дня ---')
    for citizen in citizens:
        print(citizen)
    print(my_sweet_home)
```

Блек-джек также известен как 21. Суть игры проста: либо набрать ровно 21 очко, либо набрать очков больше, чем в руках у дилера, но ни в коем случае не больше 21. Если игрок собирает больше 21, он «сгорает». В случае ничьей игрок и дилер остаются при своих.

```python
import random

class Card:
    
    def __init__(self, rank, suit):
        self.rank = rank   # ранг
        self.suit = suit   # масть
        
    def card_value(self):
        """ Возращает количество очков которое дает карта """
    
        if self.rank in "TJQK":                    # По 10 за десятку, валета, даму и короля
            return 10
        else:                                      # Возвращает нужное число очков за любую другую карту
            return " A23456789".index(self.rank)   # Туз изначально дает одно очко           
    
    def get_rank(self):
        """ Метод получения ранга карты """
        return self.rank
    
    def __str__(self):  
        return "{0} {1}".format(self.rank, self.suit)                                               
    

class Hand(object):
    """ Класс руки """
    
    def __init__(self, name):
        self.name = name       # имя игрока
        self.cards = []        # изначально рука пустая
 
    def add_card(self, card):
        """ Добавляет карту на руку """
        self.cards.append(card)
 
    def get_value(self):
        """ Метод получения числа очков на руке """
        result = 0 
        aces = 0    # количество тузов на руке
        for card in self.cards:
            result += card.card_value()
            if card.get_rank() == "A":     # если на руке есть туз - увеличиваем количество тузов
                aces += 1
        if result + aces * 10 <= 21:       # решаем считать тузы за 1 очко или за 11
            result += aces * 10
        return result
 
    def __str__(self):
        
        text = "{0}'s contains:\n".format(self.name)   
        for card in self.cards:
            text += str(card) + " "
        text += "\nHand value: " + str(self.get_value())
        return text
    
class Deck(object):
    """ Класс колоды """
    
    def __init__(self):
        ranks = "23456789TJQKA"    # ранги
        suits = "DCHS"             # масти
        self.cards = [Card(r,s) for r in ranks for s in suits]   # генератор списков создающий колоду из 52 карт
        random.shuffle(self.cards)                               # перетасовываем колоду
 
    def deal_card(self):
        """ Функция сдачи карты """
        return self.cards.pop()
    
        
def new_game():
    """ Функция, в которой реализуется процесс игры """
    
    d = Deck()   # создаем колоду
    player_hand = Hand("Player")   # задаем "руки" для игрока и компьютера
    comp_hand = Hand("Comp")
    player_hand.add_card(d.deal_card())   # сдаем две карты игроку
    player_hand.add_card(d.deal_card())
    comp_hand.add_card(d.deal_card())   # сдаем две карты компьютеру
    comp_hand.add_card(d.deal_card())
    print(comp_hand)
    print("=" * 20)
    print(player_hand)
    in_game = True    # Флаг проверки необходимости продолжать игру
    while player_hand.get_value() < 21:   # набирать карты игроку имеет смысл только если у него на руке меньше 21 очка
        ans = input("Взять карту или остановиться? Введите 1 или 2: ")
        if ans == "1":
            player_hand.add_card(d.deal_card())
            print(player_hand)
            if player_hand.get_value() > 21:   # если у игрока перебор - коапьютеру нет смысла набирать карты
                print("Вы проиграли")
                in_game = False
        else:
            print("Вы выиграли!")
            break
    print("=" * 20)
    if in_game:
        while comp_hand.get_value() < 17:   # по правилам компьютер обязан набирать карты пока его счет меньше 17
            comp_hand.add_card(d.deal_card())
            print(comp_hand)
            # Если у компьютера перебор - играть дальше нет смысла - игрок выиграл
            if comp_hand.get_value() > 21:
                print("У компьютера перебор")
                in_game = False
    if in_game: 
        if player_hand.get_value() > comp_hand.get_value():     # Ни у кого не было перебора - 
            print("Вы победили!")                               # сравниваем количество очков у компьютера и дилера
        else:
            print("Победил компьютер") # В нашей версии если у компьютера и игрока равное количество очков - выигрывает казино

new_game()
```

## 26.2 Инкапсуляция и сокрытие данных. Геттеры и сеттеры

По умолчанию атрибуты в классах являются общедоступными. Это значит, что из любого места программы мы можем получить атрибут объекта и изменить его.
```python
class Person:
    count = 0
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        Person.count += 1
        
    def __str__(self):
        return 'Имя: {}\nВозраст: {}'.format(self.name, self.age)
    
misha = Person('Миша', 20)
print(misha)
```

### Задача 1. Корабли

Даны два класса кораблей — грузовой и военный. У каждого из этих кораблей есть своя модель, и каждый может сделать два действия: сообщить свою модель и идти по воде. 

Грузовой корабль имеет такой атрибут, как заполненность, изначально он равен нулю. У него есть ещё два действия: погрузить и выгрузить груз с корабля. 

У военного же корабля нет никаких грузов, есть только оружие, которое передаётся вместе с моделью. Также, вместо погрузки и выгрузки, у него есть другое действие — атаковать.

Реализуйте классы грузового и военного кораблей. Для этого выделите общие атрибуты и методы в отдельный класс «Корабль» и используйте наследование. Не забудьте про функцию super в дочерних классах.

```python
# Пример из урока:

class Ship:
    
    def __init__(self, model):
        self.__model = model
        
    def __str__(self):
        return '\nМодель корабля: {model}'.format(model=self.__model)
    
    def sail(self):
        print('\nКорабль куда-то поплыл!')
        
class WarShip(Ship):
    def __init__(self, model, gun):
        super().__init__(model)      # вызвали то, что уже есть в базовом классе
        self.gun = gun
        
    def attack(self):
        print('\nКорабль атакует с помощью оружия', self.gun)
        
class CargoShip(Ship):
    def __init__(self, model):
        super().__init__(model)
        self.tonnage_load = 0
        
    def load(self):
        print('\nЗагружаем корабль')
        self.tonnage_load += 1
        print('Текущая загруженность:', self.tonnage_load)
        
    def unload(self):
        print('\nРазгружаем корабль')
        if self.tonnage_load > 0:
            self.tonnage_load -= 1
        else:
            print('Корабль уже разгружен!')
        print('Текущая загруженность:', self.tonnage_load)

warship = WarShip('zxc2', 'Пушки')
warship.attack()
cargoship = CargoShip('qwe3')
cargoship.load()
```

### Задача 2. Роботы

На военную базу завезли несколько интересных моделей роботов, которые похожи между собой, но имеют немного разные функции. У каждого робота есть номер модели и действие operate, которое означает, что делает робот. Особенности роботов следующие:

У робота-пылесоса есть мешок для мусора, изначально он пустой (0). При команде operate робот сообщает, что он пылесосит пол, и выводит текущую заполняемость мешка.

У военного робота есть оружие, и при команде operate он выводит сообщение о защите военного объекта с помощью этого оружия.

Ещё есть робот — подводная лодка, который также является военным. У этого робота есть значение глубины, и при команде operate он делает то же, что и военный робот, плюс сообщает, что охрана ведётся под водой.

Напишите программу, которая реализует все необходимые классы роботов.

```python
class Robot:
    def __init__(self, model):
        self.model = model
        
    def operate(self):
        print('\nРобот {} ездит по кругу'.format(self.model))
        
class Hoover(Robot):
    def __init__(self, model):
        super().__init__(model)
        self.refuse_bag = 0
        
    def operate(self):
        super().operate()
        print('Робот {} пылесосит пол'.format(self.model))
        self.refuse_bag += 1
        print('Текущая заполненность мешка:', self.refuse_bag)
        
class WarRobot(Robot):
    def __init__(self, model, gun):
        super().__init__(model)
        self.gun = gun
        
    def operate(self):
        super().operate()
        print('Робот {} защищает военный объект с помощью оружия {}'.format(self.model, self.gun))
        
class Submarine(WarRobot):
    def __init__(self, model, gun, deep):
        super().__init__(model, gun)
        self.deep = deep
        
    def operate(self):
        super().operate()  
        print('Охрана ведется под водой на глубине {} метров'.format(self.deep))

hoover = Hoover(model='Samsung')
hoover.operate()
war_robot = WarRobot(model='R2D2', gun='Пушка')
war_robot.operate()
submarine = Submarine(model='U-35', gun='Ракеты', deep=50)
submarine.operate()
```

Напишите программу, которая считывает из файла numbers.txt пары чисел, делит первое число на второе и выводит ответ на экран. Если первое число меньше второго, то программа выдаёт исключение DivisionError (нельзя делить большее на меньшее). 

Дополнительно, с помощью try except, можно обработать исключение на своё усмотрение.

```python
class DivisionError(Exception):
    pass

with open(r'task\numbers.txt', 'r') as file:
    for i_line in file:
        try:
            tmp = i_line.split()
            tmp[1] = tmp[1].rstrip('\n')
            tmp = list(map(int, tmp))
            if tmp[0] < tmp[1]:
                raise DivisionError('Нельзя делить большее на меньшее!')
            else:
                print(tmp[0] / tmp[1])
        except DivisionError:
            print('Нельзя делить большее на меньшее!')
```

Также есть два дочерних класса:

Солдат: получает урон, равный переданному значению.

Обычный гражданин: получает урон, равный двукратному переданному значению.

```python
class Unit:
    
    def __init__(self, name, hit_points):
        self.name = name
        self.hit_points = hit_points
        
    def take_damage(self, damage=0):
        """ Получить урон """
        
        self.hit_points = self.hit_points - damage
        print('{} получил урон {}'.format(self.name, damage))
        print('Состояние здоровья:', self.hit_points)
    
class Soldier(Unit):
    
    def __init__(self, name, hit_points):
        super().__init__(name, hit_points)
        
    def take_damage(self, damage):
        super().take_damage(damage)  
        
    
class Citizen(Unit):
    
    def __init__(self, name, hit_points):
        super().__init__(name, hit_points)
        
    def take_damage(self, damage):
        super().take_damage(damage * 2)
        
soldier = Soldier(name='Bob', hit_points=20)
soldier.take_damage(5)
citizen = Citizen(name='Alex', hit_points=20)
citizen.take_damage(5)
```

```python
class Property:
    """ Класс Недвижимость """
    
    def __init__(self, worth):
        self.worth = worth
        
    def calc_the_tax(self, percent):
        """ Метод расчета налога """
        return self.worth / percent
    
class Apartment(Property):
    """ Класс Квартира """
    
    def __init__(self, worth):
        super().__init__(worth)
        
    def calc_the_tax(self):
        return super().calc_the_tax(1000)

class Car(Property):
    """ Класс Машина """
    
    def __init__(self, worth):
        super().__init__(worth)
        
    def calc_the_tax(self):
        return super().calc_the_tax(200)

class CountryHouse(Property):
    """ Класс Дача """
    
    def __init__(self, worth):
        super().__init__(worth)
        
    def calc_the_tax(self):
        return super().calc_the_tax(500)

money = int(input('Введите кол-во денег: '))
price_1 = int(input('Введите стоимость квартиры: '))
price_2 = int(input('Введите стоимость машина: '))
price_3 = int(input('Введите стоимость дачи: '))
apartment = Apartment(price_1)
car = Car(price_2)
country_house = CountryHouse(price_3)
nalog = apartment.calc_the_tax() + car.calc_the_tax() + country_house.calc_the_tax()
print('Ваш налог составит:', nalog)
if money < nalog:
    print('Вам не хватает:', nalog - money)
```

Стек — это абстрактный тип данных, представляющий собой список элементов, организованных по принципу LIFO (англ. last in — first out, «последним пришёл — первым вышел»).

```python
class Fraction:
    
    """Класс для работы с дробями"""
    
    def __init__(self, a, b):
        self.a = a
        self.b = b
        
    def __str__(self):
        return '{}/{}'.format(self.a, self.b)
    
    def __add__(self, other):
        if isinstance(other, int):
            return Fraction(other * self.b + self.a, self.b) # сложение дроби с целым числом
        else: 
            return Fraction((self.a * other.b) + (other.a * self.b), (self.b * other.b)) # сложение дробей
    
    def __sub__(self, other):
        if isinstance(other, int):
            return Fraction(other * self.b - self.a, self.b)  # вычитание дроби и целого числа
        else:
            return Fraction((self.a * other.b) - (other.a * self.b), (self.b * other.b)) # вычитание дробей 
              
    def __mul__(self, other):
        if isinstance(other, int):
            return Fraction(self.a * other, self.b) # умножение дроби на число
        else:
            return Fraction(self.a * other.a, self.b * other.b) # умножение дробей
    
    def __int__(self):
        return self.a // self.b 
        
    def __float__(self):
        return self.a / self.b 
        
    
class OperationsOnFraction(Fraction):
    
    def getint(self):
        my_int = super().__int__() 
        return my_int
    
    def getfloat(self):
        my_float = super().__float__()
        return my_float
        
my_frac1 = Fraction(1, 2)
print(my_frac1) 

my_frac2 = Fraction(2, 3)
print(my_frac2)

print(my_frac1 + my_frac2)
print(my_frac1 + 2)

print(my_frac1 - my_frac2)
print(my_frac1 - 2)

print(my_frac1 * my_frac2)
print(my_frac1 * 2)

print(my_frac1.__int__())
print(my_frac1.__float__())

my_operationsonfraction = OperationsOnFraction(2, 3)
print(my_operationsonfraction)

print(my_operationsonfraction.getint())
print(my_operationsonfraction.getfloat())
```
