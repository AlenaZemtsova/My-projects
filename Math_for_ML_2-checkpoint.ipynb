В качестве практики к данному уроку найдите промежутки возрастания и убывания нескольких функций с помощью их производных (производные даны). Найдите, при каких значениях аргумента достигаются максимум и минимум как значения, при которых возрастание сменяется убыванием и наоборот. Постройте графики этих функций. Соответствуют ли полученные с помощью производных результаты графикам? 

Неравенства для производных можно решать и вручную, и с помощью sympy, решите сами, как вам удобнее. А можете решать вручную и затем проверять себя.

**Задание 1**

````python
from sympy import *
from sympy.plotting import plot
init_printing(use_unicode=False, wrap_line=False, no_global=True)
````

````python
x = Symbol('x')
````

````python
# функция:

f = x**2 + 3*x - 4
f
````

````python
# производная:

diff_f = 2*x + 3
diff_f
````

````python
# промежутки знакопостоянства (+):

solve_univariate_inequality(diff_f > 0, x)

# Производная говорит, что функция растет на промежутке от -3/2 до +∞
````

````python
# промежутки знакопостоянства (-):

solve_univariate_inequality(diff_f < 0, x)

# и убывает на промежутке от −∞ до -3/2
````

````python
# нули производной:

solve(diff_f)

# При х = -3/2 функция достигает своего min
````

````python
# посмотрим на график:

plot(f)

# все совпадает - функция убывает от −∞ до -3/2 и дальше начинает расти:
````

**Задание 1.1 - получите значение MSE меньшее 0.75**

````python
from sympy import *
from sympy.plotting import plot
init_printing(use_unicode=False, wrap_line=False, no_global=True)
import matplotlib.pyplot as plt
import numpy as np
x = Symbol('x')
````

````python
def print_points_ands_function1(sympy_function):
    def function(x_): return float(sympy_function.subs(x, x_))
    
    points_X = np.array([-2, -1, 0, 1, 2, 3, 3.5, 4, 4.5, 5])
    points_Y = np.array([3, 1, -0.5, 1.2, 2.5, 0.8, -2, -3.5, -3, -5])
    plt.xlim(-10, 10)
    plt.ylim(-5, 5)
    
    plt.scatter(points_X, points_Y, c='r')
    x_range = np.linspace(plt.xlim()[0], plt.xlim()[1], num=100)
    function_Y = [function(x_) for x_ in x_range]
    plt.plot(x_range, function_Y, 'b')
    plt.show()
    
    MSE = sum([(points_Y[i] - function(points_X[i]))**2 for i in range(len(points_Y))]) / len(points_Y)
    print(f'MSE = {MSE}')
````

````python
f1 = 2 * x * sin(2 * x + 5) + 5 * sin(2 * x + 5) + 2
f1
````

````python
print_points_ands_function1(f1)
````

````python
f2 = expand(f1.subs(x, x / 1.9))
print_points_ands_function1(f2)
````

````python
f3 = expand(f2.subs(x, x - 5))
print_points_ands_function1(f3)
````

````python
f4 = f3 - 1.6
print_points_ands_function1(f4)
````

````python
def print_points_ands_function2(sympy_function):
    def function(x_): return float(sympy_function.subs(x, x_))
    
    points_X = np.array([0, 1, 2, 3, 3.5, 4, 4.5, 5, 6, 6.5, 7, 8, 9])
    points_Y = np.array([0.5, 1.2, 0.7, 1.2, 1.1, 0.7, 0.65, 0.9, 0.95, 0.3, 0.8, 0.5, 0.0])
    plt.xlim(-1, 20)
    plt.ylim(-2, 2)
    
    plt.scatter(points_X, points_Y, c='r')
    x_range = np.linspace(plt.xlim()[0] + 1, plt.xlim()[1], num=100)
    function_Y = [function(x_) for x_ in x_range]
    plt.plot(x_range, function_Y, 'b')
    plt.show()
    
    MSE = sum([(points_Y[i] - function(points_X[i]))**2 for i in range(len(points_Y))]) / len(points_Y)
    print(f'MSE = {MSE}')
````

### 10 - Частные производные и минимизация MSE

````python
from sympy import *
````

````python
# создадим переменные:

a2, a1, a0 = symbols('a2, a1, a0')
````

````python
# Создадим функции для частных производных:

MSEa2 = 1/3 * (28 * a2 + 5000 * a1 + 12 * a0 - 5000)
MSEa1 = 1/3 * (5000 * a2 + 1090000 * a1 + 2400 * a0 - 1010000)
MSEa0 = 1/3 * (12 * a2 + 2400 * a1 + 6 * a0 - 2200)
````

````python
# распечатаем их:

MSEa2
````

````python
MSEa1
````

````python
MSEa0
````

````python
# передадим в метод nonlinsolve список уравнений и список переменных:

nonlinsolve([MSEa2, MSEa1, MSEa0], [a2, a1, a0])

# это и есть наши кандидиаты в минимумы (а2, а1 и а0):
````

````python
# Определим функцию MSE:

MSE = 1/3 * (((a2 * 2 + a1 * 200 + a0) - 200)**2 + \
             ((a2 * 1 + a1 * 450 + a0) - 300)**2 + \
             ((a2 * 3 + a1 * 550 + a0) - 600)**2)
MSE
````

````python
# и подставим в нее значения с помощью метода subs:

MSE.subs({a2: 108.333333333333, a1: 0.833333333333333, a0: -183.333333333334})
````

### 18 -  Переопределенные СЛАУ

````python
from sympy import *
import numpy
````

````python
# определим MSE:

a1, a2 = symbols('a1, a2')

MSE = 1 / 4 * ((4.1 * a1 + 3.9 * a2 - 4.0)**2 + \
               (4.7 * a1 + 4.2 * a2 - 4.3)**2 + \
               (3.1 * a1 + 2.9 * a2 - 3.0)**2 + \
               (4.9 * a1 + 5.0 * a2 - 4.9)**2)
````

````python
# и посмотрим, чему равно его значение в точке (0, 0) - эту точку мы выберем в качестве начальной:

MSE.subs({a1: 0, a2: 0})
````

````python
# теперь нам нужно найти градиент
# находим производную по а1:

MSEa1 = diff(MSE, a1)
MSEa1
````

````python
# и по а2:

MSEa2 = diff(MSE, a2)
MSEa2
````

````python
# зададим шаг:

step = 0.01
````

````python
# теперь можем задать начальную точку:

point = numpy.array([0, 0])
````

````python
# и провести первую итерацию градиентного спуска:

grad = numpy.array([MSEa1.subs({a1: point[0], a2: point[1]}), 
                    MSEa2.subs({a1: point[0], a2: point[1]})])
grad
````

````python
next_point = point - step * grad
next_point
````

````python
MSE.subs({a1: next_point[0], a2: next_point[1]})
````

### 20 - Псевдообратная матрица

````python
from sympy import *
import numpy as np
````

````python
# Рассмотрим на примере СЛАУ, которую мы решали в ДЗ к прошлому модулю
# Нам понадобилось выписать MSE, найти градиент, а затем сделать 7 итераций градиентного спуска
# прежде, чем мы достигли нужного уровня точности

# Посмотрим, что получится у нашей новой формулы
# и заодно посмотрим, как делать все пройденные в этом модуле операции с матрицами

# Для этого нам понадобится объект из sympy типа Matrix, 
# в который просто передается двумерный массив для матрицы и одномерный для вектора
# Создадим МОП для нашей системы:

X = Matrix([[ 5.0,  7.0, -5.0],
            [ 0.0, -2.0,  2.0],
            [-4.0, -8.0, -7.0],
            [ 1.0,  1.0,  2.0],
            [ 2.0, -1.0,  2.0],
            [ 4.0,  1.0,  4.0]])
X
````

````python
# и создадим вектор ответов:

b = Matrix([-47.0, 10.0, 63.0, -1.0, -4.0, -2.0])
b
````

````python
# Какие операции нам нужны, чтобы найти псевдообратную матрицу и решить задачу аппроксимации?

# Во-первых, нам нужно транспонирование:

X.T
````

````python
# Еще нам нужно умножение:

mul = X.T * X
mul
````

````python
# Теперь нам нужно найти обратную матрицу
# Для этого нам нужен детерминант (с помощью метода det)

# Как мы помним, Х не является квадратной, так что детерминанта у нее нет
# Но нам и не нужен определитель Х
# По формуле нам нужен определитель произведения Х, транспонированного на Х - а это уже квадратная матрица:

mul.det()
````

````python
# Теперь мы могли бы написать цикл, в котором легко вычислили бы все элементы матрицы алгебраических дополнений
# Но нам даже это не нужно делать, т к sympy умеет просто вычислять обратные матрицы
# Для этого используется такой же синтаксис, как просто в Python для возведения в степень:

(X.T * X)**(-1)
````

````python
# Осталось умножить на еще один Х транспонированный:

X_pseudo_inverse = (X.T * X)**(-1) * X.T
X_pseudo_inverse
````

````python
# Осталось найти ответ, умножив эту псевдообратную матрицу на вектор ответов b:

best_point = X_pseudo_inverse * b
best_point

# Это и есть наш вектор весов
````

````python
# Подставим эти веса в наше MSE из ДЗ:

x, y, z = symbols('x, y, z')

MSE = 1 / 6 * ((5 * x + 7 * y - 5 * z + 47)**2 + \
               (- 2 * y + 2 * z - 10)**2 + \
               (-4 * x - 8 * y - 7 * z - 63)**2 + \
               (x + y + 2 * z + 1)**2 + \
               (2 * x - y + 2 * z + 4)**2 + \
               (4 * x + y + 4 * z + 2)**2)
MSE.subs({x: 0, y: 0, z: 0})
````

````python
MSE.subs({x: best_point[0], y: best_point[1], z: best_point[2]})
````

````python
# Без всякого градиентного спуска и вообще без функции потерь мы в 1 проход получили результат лучше, 
# чем добились спуском за 7 итераций (там было около 55)
# Это решение и быстрее и точнее
````
