Однако во многих странах, использующих такую шкалу, эта сумма вычисляется более сложным способом: налоговая ставка 30% на доход выше 50 000 означает, что 30% уплачивается не со всей суммы, а лишь с той ее части, которая превосходит 50 000. Аналогично, ставка 20% на доход от 10 000 до 50 000 обязывает уплатить 20% лишь с той части суммы, которая превосходит 10 000, но не превосходит 50 000.

Так, например, с дохода 100 000 придется заплатить такой налог:

30% * (100 000 − 50 000) + 20% * (50 000 − 10 000) + 13% * 10 000 = 15 000 + 8 000 + 1 300 = 24 300

А с дохода 30 000 — такой:  20% * (30 000 − 10 000) + 13% * 10 000 = 4 000 + 1 300 = 5 300

Напишите программу, которая спрашивает у пользователя его доход и высчитывает сумму налога для него по вышеописанным правилам.

```python
# Задача №5 из homework3_jun_anl_loops_and_ifs

# Простое число делится только на 1 и на себя (2, 3, 5, 7, 11, 13, 17 и т д)
# Составное число делится на 1, на себя и на другие числа (6, 8, 9, 10, 12, 14 и т д)
# число 1 - не простое и не составное

number = int(input('Введите число: '))
isPrime = True
for divider in range(2, number):
    if number % divider == 0:   # если остаток равен нулю, мы нашли делитель - значит число составное
        isPrime = False
        break
if isPrime:
    print('Число простое')
else:
    print('Число составное')
```

N! = 1 * 2 * 3 * 4 * 5 * … * N

Пример:

```python
n = int(input('Введите число: '))
factorial = 1
for i in range(1, n + 1):
    if n == 1:
        print('Факториал числа', n, 'равен', 1)
    elif n == 0:
        print('Факториал числа', n, 'равен', 1)
    else:
        factorial *= i
print('Факториал числа', n, 'равен', factorial)
```

```python
# Последовательность нечетных чисел от 1 до N: 1,3,5,7,9...
# Каждое из них больше предыдущего на 2
# Решим с помощью цикла while:

n = int(input('Введите число: '))
number = 1
while number <= n:
    print(number, '** 2 =', number ** 2)
    number += 2
```

```python
# Так как чётное число - это то число, которое делится на 2, то его можно записать, как 2n, где n — целое число, 
# а нечётное — соответственно как 2n+1:

n = int(input('Введите число: '))
for i in range(1, n//2 + n % 2 + 1):
    i = i * 2 - 1
    print(i, '** 2 =', i ** 2)
```

Два сотрудника спецслужб переписываются необычным шифром. Каждую букву они шифруют в виде строки, внутри которой есть длинная последовательность букв “s”, а длина самой длинной - это и есть номер буквы алфавита, которую хотят отправить.

Напишите программу, которая получает на вход строку, подсчитывает в ней самую длинную последовательность подряд идущих букв “s” и выводит ответ на экран.

Пример:

Введите строку: ssbbbsssbc

Самая длинная последовательность: 3

```python
# 1-й вариант:

string = input('Введите строку: ')
max_equal = 0
max_max = 0

for i in range(len(string)):
    if max_equal > 0:
        if string[i] == 's' and string[i - 1] == 's':
            max_equal += 1
        else:
            max_equal = 0
            
    elif string[i] == 's':
        max_equal += 1
    if max_equal > max_max:
        max_max = max_equal
        
print('Самая длинная последовательность:', max_max)
```

```python
# 2-й вариант:

string = input('Введите строку: ')
curr_seq = 0
max_seq = 0

for sym in string:
    if sym == 's':
        curr_seq += 1
    else:
        curr_seq = 0
    if curr_seq > max_seq:
        max_seq = curr_seq
        
print('Самая длинная последовательность:', max_seq)
```

**166 - Наибольшая сумма цифр**

Вводится N чисел. Среди натуральных чисел, которые были введены, найдите наибольшее по сумме цифр. Выведите на экран это число и сумму его цифр.

```python
max_summ = 0    # сумма цифр числа с наибольшей суммой цифр
max_number = 0  # число с наибольшей суммой цифр

n = int(input('Введите кол-во чисел: '))
for i in range(1, n + 1):
    number = int(input(f'Введите {i} число: '))
    
    summ = 0   # сумма текущего числа
    temp = number  # записываем в temp текущее значение number
    while temp > 0:
        summ += temp % 10  # прибавляем тройку
        temp //= 10        # отбрасываем тройку
    if summ > max_summ:
        max_summ = summ
        max_number = number
        
print('Число', max_number, 'имеет максимальную сумму цифр:', max_summ)
```

```python
# получаем тройку:

123 % 10
```

```python
# прибавили тройку к summ
# отбросили тройку:

123 // 10
```

```python
# получаем двойку:

12 % 10
```

```python
# прибавили двойку к summ
# отбросили двойку:

12 // 10
```

```python
# получаем единицу:

1 % 10
```

```python
# прибавили единицу к summ
# отбросили единицу:

1 // 10

# дошли до нуля, цикл закончился
```

**167 - Пирамидка**

Напишите программу, которая выводит на экран равнобедренный треугольник (пирамидку), заполненный символами хэштега #. Пусть высота пирамиды вводится пользователем.

Подсказка: вспомните, как выводился колонтитул вида -----!!!!!!-----

![title](Циклы21.jpg)

```python
height = int(input('Введите высоту пирамиды: '))
print()
hashtags = 1

for line in range(height):
    print(' ' * (height - line - 1), end = '')
    print('#' * hashtags)
    hashtags += 2
```

```python
# Считаем кол-во пробелов по формуле: height - line - 1 (высота - строка - 1)

# высота 5 - линия 0 - 1 = 4 пробела
# высота 5 - линия 1 - 1 = 3 пробела
# высота 5 - линия 2 - 1 = 2 пробела
# высота 5 - линия 3 - 1 = 1 пробел
# высота 5 - линия 4 - 1 = 0 пробелов

# end = '' приклеиваем следующий принт с хэштегами

# Хэштеги берем сначала один, потом каждый раз прибавляем по два
```

**169 - Яма**

В одной компьютерной текстовой игре рисуются всяческие элементы ландшафта.

Напишите программу, которая получает на вход число N и выводит на экран числа в виде «ямы» вот так: 

![title](Циклы20.jpg)

```python
depth = int(input('Введите глубину ямы: '))

for line in range(depth):
    for left_number in range(depth, depth - line - 1, -1):
        print(left_number, end = '')
        
    point_count = 2 * (depth - line - 1)
    print('.' * point_count, end = '')
    
    for right_number in range(depth - line, depth + 1):
        print(right_number, end = '')
    print()
```
