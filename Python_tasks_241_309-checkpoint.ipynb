В базе данных небольшой библиотеки хранятся ID выданных книг. С утра выдали всего 7 штук. Правда парочку из них уже в это утро вернули обратно. Поэтому в базе они обозначаются числом -1. В течение дня посетители взяли так же еще 10 книг, а некоторые из них тоже вернули обратно. Нужно написать программу, которая на утро следующего дня сгенерирует нам новый список из ID только выданных книг, а также посчитает, сколько книг вернули.

```python
books_ID = [50, 34, -1, -1, 2, 23, -1]
new_books_ID = []
returned = 0

for _ in range(10):
    id = int(input('Введите ID книги: '))
    books_ID.append(id)
    
for id in books_ID:
    if id == -1:
        returned += 1
    else:
        new_books_ID.append(id)
        
print('Новый список выданных книг:', new_books_ID)
print('Вернули за день:', returned)
```

#### 246 - Урон

У нас есть список из n элементов - там хранятся значения урона от всех монстров. Среди этих монстров есть один такой маг, который берет свое значение урона и добавляет его ко всем остальным, которые меньше 100. Кроме себя, конечно же.

```python
monsters_count = int(input('Кол-во монстров: '))
mage_index = int(input('Номер мага в списке: '))
monsters_damage = []

for monster in range(monsters_count):
    print('Урон у', monster + 1, 'монстра:', end = ' ')
    damage = int(input(''))
    monsters_damage.append(damage)
    
for i in range(monsters_count):
    if monsters_damage[i] < 100 and i != mage_index - 1:
        monsters_damage[i] += monsters_damage[mage_index - 1]
        
print('Итоговый урон монстров:', monsters_damage)
```

#### 248 - Кратность

Пользователь вводит список из N чисел и число K. Напишите код, выводящий на экран сумму индексов элементов списка, которые кратны K.

Пример:

Кол-во чисел в списке: 4

Введите 1 число: 1

Введите 2 число: 20

Введите 3 число: 30

Введите 4 число: 4


Введите делитель: 10


Индекс числа 20: 1

Индекс числа 30: 2

Сумма индексов: 3

```python
n = int(input('Кол-во чисел в списке: '))
numbers = []
summ = 0

for i in range(n):
    print('Введите', i + 1, 'число:', end = ' ')
    number = int(input())
    numbers.append(number)
    
k = int(input('Введите делитель: '))

for i in range(n):
    if numbers[i] % k == 0:
        print('Индекс числа', str(numbers[i]) + ':', i)
        summ += i
    else:
        i += 1
        
print('Сумма индексов:', summ)
```

#### 249 - Собачьи бега

В собачьих бегах участвует N собак, у каждой из них есть определённое количество очков за сезон. На огромном табло выводятся очки каждой собаки. Однако при выводе был обнаружен баг: собаки с наибольшим и наименьшим количеством очков поменялись местами! Нужно это исправить.

Дан список очков из N собак. Напишите программу, которая меняет местами наибольший и наименьший элементы в списке.

```python
n = int(input('Введите кол-во собак: '))
scores = []

for i in range(n):
    score = int(input('Введите очки: '))
    scores.append(score)

maxx = 0
minn = 1000000
i_maxx = 0
i_minn = 0
for j in range(len(scores)):
    if scores[j] > maxx:
        maxx = scores[j]
        i_maxx = j
    if scores[j] < minn:
        minn = scores[j]
        i_minn = j

scores[i_maxx] = minn
scores[i_minn] = maxx
        
print('Список очков собак:', scores)
```

В этот раз у нас стоит задача сделать фишку с поиском и заменой символов в выделенной строчке. Например, человек что-то перечислял в тексте, но ошибся и вместо точек с запятой использовал двоеточия. Лингвисты негодуют.

Пользователь вводит строку S. Напишите программу, которая заменяет в строке все двоеточия (:) на точки с запятой (;). Также подсчитайте количество замен и выведите ответ на экран (и новую строку тоже). Для решения используйте список.

Пример:

Введите строку: гвозди:шурупы:гайки

Исправленная строка: гвозди;шурупы;гайки

Кол-во замен: 2

```python
st = input('Введите строку: ')
sym = ':'
replace_sym = ';'
summ = 0

st = list(st)

for i in range(len(st)):
    if st[i] == sym:
        st[i] = replace_sym 
        summ += 1

print('Исправленная строка: ', end = '')
for i in st:
    print(i, end = '')

print()
print('Кол-во замен:', summ)
```

Пример 1:

Введите строку: abbc

Номер символа: 3

Символ слева: b

Символ справа: c

Есть ровно один такой же символ.

Пример 2:

Введите строку: abсd

Номер символа: 3

Символ слева: b

Символ справа: d

Таких же символов нет.

```python
st = input('Введите строку: ')
sym_num = int(input('Номер символа: '))

my_sym = st[sym_num - 1]
sym_left = st[sym_num - 2]
sym_right = st[sym_num]

print('Символ слева:', sym_left)
print('Символ справа:', sym_right)

if my_sym == sym_left and my_sym == sym_right:
    print('Есть ровно два таких же символа.')
elif my_sym == sym_left or my_sym == sym_right:
    print('Есть ровно один такой же символ.')
else:
    print('Таких же символов нет.')
```

Затем вводится сам текст произведения строго по словам. Текст вводится до тех пор, пока не встретится слово end. Напишите программу, которая посчитает, сколько раз слова пользователя встречаются в тексте.

Пример:

Введите 1 слово: я

Введите 2 слово: год

Введите 3 слово: лучший


Слово из текста: этот

Слово из текста: год

Слово из текста: -

Слово из текста: лучший

Слово из текста: год


Подсчёт слов в тексте

я: 0

год: 2

лучший: 1

```python
words_list = []
counts = [0, 0, 0]

for i in range(3):
    print('Введите', i + 1, 'слово:', end = ' ')
    word = input()
    words_list.append(word)
    
text = input('Слово из текста: ')
while text != 'end':
    for index in range(3):
        if words_list[index] == text:
            counts[index] += 1
    text = input('Слово из текста: ')
    
print('\nПодсчет слов в тексте')
for i in range(3):
    print(words_list[i], ':', counts[i])
```

```python
word = input('Введите слово: ')
syms_list = []

for i_sym in word:
    count_char = 0
    for y_sym in word:
        if i_sym == y_sym:
            count_char += 1
        if count_char > 1:
            break
    if count_char == 1:
        syms_list.append(i_sym)
           
print('Кол-во уникальных букв:', len(syms_list))
print('Сами буквы:', *syms_list)
```

Дан список из N элементов и целое число K. Напишите программу, которая циклически сдвигает элементы списка вправо на K позиций. Используйте минимально возможное количество операций присваивания.

Пример 1:

Сдвиг: 1

Изначальный список: [1, 2, 3, 4, 5]

Сдвинутый список: [5, 1, 2, 3, 4]

Пример 2:

Сдвиг: 3

Изначальный список: [1, 4, -3, 0, 10]

Сдвинутый список: [-3, 0, 10, 1, 4]

```python
initial_list = [1, 4, -3, 0, 10]
shift_list = [1, 4, -3, 0, 10]

shift = int(input('Сдвиг: '))

for i in range(len(initial_list)):
    if i + shift >= len(initial_list):
        shift_list[i - len(initial_list) + shift] = initial_list[i]
    else:
        shift_list[i + shift] = initial_list[i]

print('Изначальный список:', initial_list)
print('Сдвинутый список:', shift_list)
```

#### 264 - Сортировка

Постарайтесь придумать и написать как можно более эффективный алгоритм сортировки.

Пример:

Изначальный список: [1, 4, -3, 0, 10]

Отсортированный список: [-3, 0, 1, 4, 10]

```python
# Сортировка выбором:

nums = [4, 9, 7, 6, 3, 2]
print('Изначальный список:', nums)

def selection_sort(my_list):
    
    for i_mn in range(len(my_list)):                 # цикл для попарного сравнения элементов
        for curr in range(i_mn, len(my_list)):
            if my_list[curr] < my_list[i_mn]:
                my_list[curr], my_list[i_mn] = my_list[i_mn], my_list[curr]

selection_sort(nums)

print('Отсортированный список:', nums)

# i_mn - сокращение от minimum
```

```python
# Задача "Улучшенная лингвистика" ("Подсчет слов")

words_list = []
counts = [0, 0, 0]

for i in range(3):
    print('Введите', i + 1, 'слово:', end = ' ')
    word = input()
    words_list.append(word)
    
text = input('Слово из текста: ')
while text != 'end':
    for index in range(3):
        if words_list[index] == text:
            counts[index] += 1
    text = input('Слово из текста: ')
    
print('\nПодсчет слов в тексте')
for i in range(3):
    print(words_list[i], ':', counts[i])
```

Два класса стоят в две отдельные шеренги. В каждом классе ученики выстроены по росту (по возрастанию): в одном классе от 160 см до 176 см с шагом 2, во втором классе — от 162 см до 180 см с шагом 3. Спустя какое-то время два класса объединяют в одну шеренгу и тоже выстраивают их по возрастанию.

Напишите программу, которая генерирует списки роста для каждого в классе, затем объединяет их в один список и сортирует его в порядке возрастания. Выведите отсортированный список на экран.

```python
def selection_sort(my_list):
    
    for i_mn in range(len(my_list)):                 
        for curr in range(i_mn, len(my_list)):
            if my_list[curr] < my_list[i_mn]:
                my_list[curr], my_list[i_mn] = my_list[i_mn], my_list[curr]

class_1 = []
class_2 = []

for i in range(160, 177, 2):
    class_1.append(i)
    
for j in range(162, 181, 3):
    class_2.append(j)
    
print('Первый класс:', class_1)
print('Второй класс:', class_2)

class_1.extend(class_2)
selection_sort(class_1)
print('Шеренга из двух классов по возрастанию:', class_1)
```

#### 286 - Симметричная последовательность (Палиндром с цифрами)

Последовательность чисел называется симметричной, если она одинаково читается как слева направо, так и справа налево. Например, следующие последовательности являются симметричными:

**Что-то подобное было в задаче "Палиндром", только здесь вместо букв - цифры:**

![title](Списки4.jpg)

**Пример 2** - если то, что мы перевернули добавить в исходный список чисел, получится правильно. Но нам надо взять **минимальное кол-во чисел**, а это значит, что **двойка по центру тут явно лишняя**. Если ее убрать, то тоже получится палиндром. Значит **надо придумать, как ее не брать**. И здесь явно одним только переворачиванием дело не обходится. Посмотрим на это с другой стороны:

![title](Списки5.jpg)

```python
# Здесь мы могли бы использовать метод списков reverse, который сам переворачивает список.
# Однако мы передаем в качестве аргумента список (num_list) и мы помним, что в таком случае список изменится
# Поэтому мы пойдем классическим путем - через новую переменную и цикл:

def is_palindrome(num_list):
    reverse_list = []
    for i_num in range(len(num_list) -1, -1, -1):
        reverse_list.append(num_list[i_num])
    if num_list == reverse_list:
        return True
    else:
        return False

nums = [1, 2, 3, 4, 3, 2]  # для удобства возьмем готовый список и для разных тестов будем его менять

# Мы знаем, что нам нужно будет каждый раз заново брать элементы списка, начиная с определенного индекса
# А также нам нужно будет где-то хранить ответ
# Заведем для этого еще 2 переменные:
new_nums = []
answer = []

# Теперь нам нужен цикл для прохода по нашему списку:
for i_nums in range(0, len(nums)):
    
    # дальше нам понадобится каждый раз заполнять новый список из списка nums
    # причем начиная не с нулевого элемента, а именно с текущего:
    for j_elem in range(i_nums, len(nums)):
        new_nums.append(nums[j_elem])
        
    # После этого нам нужно проверить, не является ли наш новый список палиндромом
    # Сделаем это с помощью вызова функции:
    if is_palindrome(new_nums):
        
        # если текущая посл-ть палиндром, то мы все уже нашли
        # нужно только записать в ответ те числа, которые стоят до того, на котором мы остановились в основном цикле
        for i_answer in range(0, i_nums):
            answer.append(nums[i_answer])
            
        # нужно также не забыть этот список перевернуть и выйти из циклов
        # здесь уже можно спокойно воспользоваться методом reverse():
        answer.reverse()
        break
        
    # если посл-ть не оказалась палиндромом, то мы идем заново, а значит нужно переприсвоить новый список:
    new_nums = []  
    
print('Последовательность:', nums)
print('Нужно приписать чисел:', len(answer))
print('Сами числа:', answer)       
```

Напишите программу, которая получает на вход список цен на товары (вещественные числа, список генерируется также с помощью list comprehensions) и выводит в одну строку общую сумму стоимости товаров за каждый год.

Пример: 

Цена на товар: 1.09

Цена на товар: 23.56

Цена на товар: 57.84

Повышение на первый год: 0

Повышение на второй год: 10

Сумма цен за каждый год: 93.83 93.83 103.22

```python
def get_higher_price(percent, price):
    return round(price * (1 + percent / 100), 2)

prices_now = [1.09, 23.56, 57.84, 4.56, 6.78]

first_percent = int(input('Повышение на первый год: '))
second_percent = int(input('Повышение на второй год: '))

prices_first = [get_higher_price(first_percent, i_price) for i_price in prices_now]
prices_second = [get_higher_price(second_percent, i_price) for i_price in prices_first]

print('Сумма цен за каждый год:', round(sum(prices_now), 2), round(sum(prices_first), 2), round(sum(prices_second), 2))
```

#### 294 - Клиент - сервер

Мы тестируем клиент-серверную систему у себя на компьютере. Сервер генерирует список из чисел, которые делятся на 10, начиная с 1 и доходя до 100. Нам нельзя менять данные на сервере напрямую, поэтому нам нужно взять копию этого списка, занулить элемент с индексом 3 и вывести на экран элементы с индексами со 2-го по 7-й.

```python
nums = [x for x in range(1, 101) if x % 10 == 0]
new_nums = nums
new_nums[3] = 0

for i_elem in range(2, 8):
    print(nums[i_elem], end=' ')
    
print()

for i_elem in range(2, 8):
    print(new_nums[i_elem], end=' ')
```

Программа вывела нам одинаковые ответы, хотя внутри циклов мы используем разные списки. На самом деле, **это не два списка, а один**. Мы берем просто еще одну переменную new_nums, которая ссылается на точно такой же список. И теперь получается, что у одного списка у нас два названия. Так что **копию мы здесь не делаем**. И чтобы это исправить, **сделаем так**:

```python
nums = [x for x in range(1, 101) if x % 10 == 0]
new_nums = nums[:]  # добавили срез [:]
new_nums[3] = 0

for i_elem in range(2, 8):
    print(nums[i_elem], end=' ')
    
print()

for i_elem in range(2, 8):
    print(new_nums[i_elem], end=' ')
```

Теперь мы получили копию списка. По условию задачи, нам нужно вывести на экран элементы с индексами со 2-го по 7-й:

```python
nums = [x for x in range(1, 101) if x % 10 == 0]
new_nums = nums[:]  
new_nums[3] = 0

print(new_nums[2:8])
```

#### 307 - Развлечение

N палочек выставили в один ряд, пронумеровав их слева направо числами от 1 до N. Затем по этому ряду бросили K камней, при этом i-й камень сбил все палки с номерами от L_i до R_i включительно. Определите, какие палки остались стоять на месте.

Напишите программу, которая получает на вход количество палок N и количество бросков K. Далее идёт K пар чисел L_i, R_i, при этом 1≤ L_i≤ R_i≤ N.

Программа должна вывести последовательность из N символов, где j-й символ есть “I”, если j-я палка осталась стоять, или “.”, если j-я палка была сбита.

Пример:

Кол-во палок: 10 

Кол-во бросков: 3

Бросок 1. Сбиты палки с номера 8 

по номер 10

Бросок 2. Сбиты палки с номера 2 

по номер 5

Бросок 3. Сбиты палки с номера 3 

по номер 6

Результат: I.....I...

```python
n = int(input('Кол-во палок: '))
k = int(input('Кол-во бросков: '))
res = ['|' for i in range(n)]

for i in range(1, k + 1):
    L_i = int(input(f'Бросок {i} . Сбиты палки с номера '))
    R_i = int(input('по номер '))
    
    for j in range(L_i, R_i + 1):
        res[j - 1] = '.'

print('Результат: ', end='')
for i in res:
    print(i, end='')
    
# Можно так: print('Результат: ', *res, sep='')
```

#### 309 - Шифр Цезаря

Юлий Цезарь использовал свой способ шифрования текста. Каждая буква заменялась на следующую по алфавиту через K позиций по кругу. Если взять русский алфавит и k = 3, то в слове, которое мы хотим зашифровать, буква А станет буквой Г, Б станет Д и так далее.

Пользователь вводит сообщение, а также значение сдвига. Напишите программу, которая зашифрует это сообщение при помощи шифра Цезаря.

Пример:

Введите сообщение: это питон

Введите сдвиг: 3

Зашифрованное сообщение: ахс тлхср

```python
def caesar_cipher(string, user_shift):
    
    char_list = [(alphabet[(alphabet.index(sym) + user_shift) % 33] if sym != ' ' else ' ') for sym in string]
    new_str = ''
    for i_char in char_list:
        new_str += i_char
    return new_str

alphabet = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'

input_str = input('Введите сообщение: ')
shift = int(input('Введите сдвиг: '))

output_str = caesar_cipher(input_str, shift)
    
print('Зашифрованное сообщение:', output_str)
```
